```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Plotting a vector

```python
plt.quiver(0, 0, 4, 5)
plt.show()
```

```python
plt.quiver(0, 0 , 4, 5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0 , 4, 5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0 , 4, 5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0 , 3, -6, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0, 0 , 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0 , 3, -2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0 , 5, 1, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Substraction for two vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])
diff = vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0, 0 , 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0 , 3, -2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0 , -1, 5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

Multiplying a vector by scalar

```python
vector_1 = np.asarray([0, 0, 2, 3])
mul = 2 * vector_1
print(mul)
```

```python
plt.quiver(0, 0 , 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0 , 4, 6, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
vector_1 = np.asarray([0, 0, 2, 3])
mul = -0.5 * vector_1
print(mul)
```

```python
plt.quiver(0, 0 , 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(-0, -0 , -1, -1.5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
import numpy as np
```

Dot product of two vectors

```python
a  = np.array([2, 3])
b = np.array([4, 4])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

Cross product of two vectors

```python
a  = np.array([2, 3])
b = np.array([4, 4])
a_cross_b = np.cross(a, b)
print(a_cross_b)
```

Projection of one vector on another

```python
a = np.array([2, 5])
v = np.array([8, -6])
#magnitude of v vector
magnitude_of_v = np.sqrt(np.sum(v**2))
projection_of_a_on_v  = (np.dot(a, v)/magnitude_of_v**2)*v
print("projection of a vector on v vector = ", projection_of_a_on_v)
```
