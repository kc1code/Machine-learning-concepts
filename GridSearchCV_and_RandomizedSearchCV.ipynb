```python
#importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

taking breast cancer dataset

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
#printinf the first 5 rows
data_frame.head()
```

```python
#adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
 #checking the distribution of target variable
 data_frame['label'].value_counts()
```

separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

#**GridSearchCV**

GridSearchCV is used to determining the best parameter for our model

```python
#loding th SVC model
model = SVC()
```

```python
#hyperparameters
parameters = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'], 'C':[1,5,10,20]}
```

```python
#grid search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
#fitiing the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
#best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}

#**RandomizedSearchCV**

```python
#loding th SVC model
model = SVC()
```

```python
#hyperparameters
parameters = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'], 'C':[1,5,10,20]}
```

```python
#grid search
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
#fitiing the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
#best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
#loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
randomized_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}
